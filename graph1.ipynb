```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-10,10,100)
y=x
plt.subplot(2,1,1)
plt.plot(x,y,'g--')
plt.title("identity")
plt.subplot(2,1,2)
y1=[]
i=0
while(i<len(x)):
    if(x[i]>=1):
        y1.append(1)
    if(x[i]<1):
        y1.append(0)
    i=i+1
plt.plot(x,y1,'b--')
plt.title("binary")
plt.show()
```
